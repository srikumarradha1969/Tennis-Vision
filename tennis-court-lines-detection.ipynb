```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import time
import imutils
```

```python
cap = cv2.VideoCapture("VideoInput/video_cut.mp4")
fps = int(cap.get(cv2.CAP_PROP_FPS))
print('fps : {}'.format(fps))
kernel = np.ones((2,2),np.uint8)
# try to determine the total number of frames in the video file
if imutils.is_cv2() is True :
    prop = cv2.cv.CV_CAP_PROP_FRAME_COUNT
else : 
    prop = cv2.CAP_PROP_FRAME_COUNT
total = int(cap.get(prop))
# start from first frame
currentFrame = 0
```

```python
# Define the codec and create VideoWriter object  
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
print(frame_width, " ",frame_height )
fourcc = cv2.VideoWriter_fourcc('M', 'J', 'P', 'G')
out = cv2.VideoWriter("VideoOutput/video_cut_output.avi", fourcc, fps,(frame_width,frame_height),True )
```

```python
while(cap.isOpened()):  

    print('Percentage of video processed : {}'.format(round( (currentFrame / total) * 100, 2)))   
    # capture frame-by-frame
    cap.set(1, currentFrame);
    
    ret, frame = cap.read()  
    if ret==True:  
        
        #Convert image to grayscale and Gaussian blur
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # To retain the original image
        RGB_img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # Inverse thresholding to get distinct white lines
        thresh = cv2.threshold(gray, 160, 255, cv2.THRESH_BINARY)[1] # increase threshold to 200 for input_video_1a.mp4
        # Try to increase intensity of the paler white lines
        # Trying Morphological Operations
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
       
        dilated = cv2.dilate(thresh, kernel, iterations=1)
        
        # find contours
        contours = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)[0]
        
        
        selected_contours = []
        
        for contour in contours:
            if cv2.contourArea(contour) > 7000:
                selected_contours.append(contour)
        
        # Sorting out contours based on area from highest to lower
        contours1 = sorted(selected_contours, key=cv2.contourArea,reverse=True)
        
        # Trying to filter to closed contours
        closed_contours = []
        open_contours = []
        for i in contours1:
            if cv2.contourArea(i) > cv2.arcLength(i, True):
                closed_contours.append(i)
            else:
                open_contours.append(i)
        
        # Check if Centroid lies within the maximum contour
        closed_countoursyes = []
        closed_countoursno = []
        for j in closed_contours:
            M = cv2.moments(j)
            cx = int(M['m10']/M['m00'])
            cy = int(M['m01']/M['m00'])
            result = cv2.pointPolygonTest(closed_contours[0], (cx ,cy), False) 
            
            if result==1.0:
                closed_countoursyes.append(j)
            else:
                closed_countoursno.append(j)
        
        # Drawing the closed contours on the frame
        finalimage = cv2.drawContours(frame, closed_countoursyes, -1, (255,0,0), thickness=cv2.FILLED)
        
        out.write(finalimage)
        
        # next frame
        currentFrame += 1
        # Display the resulting frame
        cv2.imshow('frame',finalimage)
                            
        if cv2.waitKey(1) & 0xFF == ord('q'):
               break
    else:  
        break  
  
# Release everything if job is finished  
cap.release()  
out.release()  
cv2.destroyAllWindows()  
```
